```python
import numpy as np 
import seaborn as sns
import pandas as pd 
from sklearn.datasets import load_wine
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
#load the dataset from load_digits
digits = load_digits()
```

```python
data = digits.data
feature_name = digits.feature_names
target = digits.target
target_name = digits.target_names
```

# Using TSNE on MNIST Dataset


```python


#standarize the data using sklearn standardscaler 
scaler = StandardScaler()
standard_data = scaler.fit_transform(data)
#this will get the data down to the same scale so that it will be easy to work on it 


#use tsne on the transformed data 
tsne = TSNE(n_components=2, random_state = 0 , perplexity=3 , n_iter = 1000)
#n_compnents is the no of dimesnion of the embedding 
#random_state is used to produce same output across multiple runs 
#preplexity = number of neigbours that has to consider for the embedding 
#n_iter is the max number of itteration for which tsne will work 

transformed_data = tsne.fit_transform(standard_data)
#transform the standardize data into 2 dim 

#adding label to the data after dim reduction and then convert it into pandas dataframe object 
tsne_data = np.vstack((transformed_data.T , target)).T
df = pd.DataFrame(tsne_data , columns=["dim1" , "dim2" , "labels"])


plt.figure(figsize=(10, 10))
sns.FacetGrid(df, hue="labels", height=6).map(plt.scatter, 'dim1', 'dim2').add_legend()
plt.show()



```

# creating simple model using if else condition


```python
def model(dim1 , dim2):
    if (dim2 < -5 ):
        return 2
    elif (dim2 < 4 and dim2 > -5):
        return 1
    else:
        return 0

df['predicted'] = df.apply(lambda x : model(x['dim1'] , x['dim2']) , axis=1)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(df['labels'] , df['predicted'])
```
